```python
!pip install ultralytics --quiet
from pathlib import Path
from PIL import Image
import random, shutil, os
from tqdm import tqdm

base_dir = Path("/content/rps_yolo")
images_dir = base_dir / "images"
labels_dir = base_dir / "labels"

for split in ['train', 'val']:
    (images_dir / split).mkdir(parents=True, exist_ok=True)
    (labels_dir / split).mkdir(parents=True, exist_ok=True)

# íƒìƒ‰ ê²½ë¡œ
uploaded_path = Path("/content/rps_raw")
target_classes = ['rock', 'paper', 'scissors']
found_folders = []

# ëª¨ë“  í•˜ìœ„ í´ë”ì—ì„œ rock/paper/scissors íƒìƒ‰
for root, dirs, files in os.walk(uploaded_path):
    for d in dirs:
        if d in target_classes:
            found_folders.append(Path(root) / d)

# ë¼ë²¨ ìƒì„± í•¨ìˆ˜
def create_label(img_path, label_path, class_id):
    img = Image.open(img_path)
    width, height = img.size
    x_center, y_center = 0.5, 0.5
    bbox_width, bbox_height = 1.0, 1.0
    label = f"{class_id} {x_center} {y_center} {bbox_width} {bbox_height}\n"
    with open(label_path, 'w') as f:
        f.write(label)

# ì´ë¯¸ì§€/ë¼ë²¨ ì €ì¥
class_map = {name: idx for idx, name in enumerate(target_classes)}

for folder in found_folders:
    cls = folder.name
    class_id = class_map[cls]
    images = list(folder.glob("*.png")) + list(folder.glob("*.jpg")) + list(folder.glob("*.jpeg"))
    random.shuffle(images)
    split_idx = int(len(images) * 0.8)

    for i, img_path in enumerate(tqdm(images, desc=f"{cls}")):
        split = "train" if i < split_idx else "val"
        img_dst = images_dir / split / img_path.name
        lbl_dst = labels_dir / split / f"{img_path.stem}.txt"
        shutil.copy(img_path, img_dst)
        create_label(img_path, lbl_dst, class_id)

# data.yaml ìƒì„±
yaml_path = base_dir / "data.yaml"
yaml_content = f"""
path: {base_dir}
train: images/train
val: images/val
names:
"""

for i, name in enumerate(target_classes):
    yaml_content += f"  {i}: {name}\n"

with open(yaml_path, "w") as f:
    f.write(yaml_content)

print("âœ… YOLO í˜•ì‹ìœ¼ë¡œ ë³€í™˜ ì™„ë£Œ!")
```

```python
import os
import shutil

val_dir = "/content/archive/Rock-Paper-Scissors/val"
classes = ['rock', 'paper', 'scissors']

# í´ë˜ìŠ¤ë³„ í´ë” ìƒì„±
for cls in classes:
    os.makedirs(os.path.join(val_dir, cls), exist_ok=True)

# íŒŒì¼ë“¤ì„ í´ë˜ìŠ¤ ê¸°ì¤€ìœ¼ë¡œ í´ë”ë¡œ ì´ë™
for filename in os.listdir(val_dir):
    filepath = os.path.join(val_dir, filename)
    if os.path.isfile(filepath):
        filename_lower = filename.lower()
        for cls in classes:
            if cls in filename_lower:
                shutil.move(filepath, os.path.join(val_dir, cls, filename))
                break
```
